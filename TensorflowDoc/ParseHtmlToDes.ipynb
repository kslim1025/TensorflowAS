```python
from bs4 import BeautifulSoup
import os
import json
from urllib import parse

def readXml(filepath):
    return BeautifulSoup(open(filepath,encoding="utf8"))

def traceChildren(children):
    for child in children:
        print("~~~",child)
def traceObj(obj):
    print("traceObj:",obj)
    for key in obj:
        print(key)

def getNodeClass(node):
    try:
        return node["class"]
    except:
        return None;



def saveNavObj(navlist):
    navO={};
    navO["Nav"]=navlist
    fw=open("website/"+"nav"+".json","w",encoding="utf-8")
    json.dump(navO,fw,ensure_ascii=False,indent=4)
def getSimpleText(node):
    rst=node.string
    if rst==None:
        s=node.stripped_strings
        sl=[]
        for st in s:
            sl.append(st)
        rst=" ".join(sl)
    return rst
def parseParamO(param):
    paramName=param.find(None,"param-name").string
    paramType=getSimpleText(param.find(None,"param-type"))
    paramDoc=getSimpleText(param.find(None,"param-docs"))
    
    #print(paramName,paramType,"@",paramDoc)
    paramO={}
    paramO["name"]=paramName;
    paramO["type"]=paramType
    paramO["doc"]=paramDoc
    return paramO
    
def parseFunctionO(funO):
    funData={}
    funName=funO.find(None,"symbol-link").string.strip()
    print(funName)
    params=funO.find_all(None,"parameter")
    funData["name"]=funName
    paramList=[]
    
    if params!=None:
        print("params:",len(params))
        for param in params:
            tclz=getNodeClass(param)
            if len(tclz)>1:
                #print(tclz)
                continue
            
            paramList.append(parseParamO(param))
    funData["params"]=paramList
    
    returns =funO.find(None,"return-type")
    if returns!=None:
        funData["return"]=getSimpleText(returns)
        #print("return:",getSimpleText(returns))
        
    return funData

def workFunctionList(funList):
    rst=[]
    for funO in funList:
        rst.append(parseFunctionO(funO))
        
    return rst

def parseClassO(classO):
    rstO={}
    clzName=classO.find(None,"symbol-link").string.strip()
    print("class:",clzName)
    sign=classO.find(None,"signature")
    if sign!=None:
        pclz=getSimpleText(sign)
        if "extends" in pclz:
            pclz=pclz.replace('extends',"")
            pclz=pclz.strip()
            rstO["extends"]=pclz
        print("sign:",pclz)
    funs=classO.find_all(None,"symbol function method")
    fList=[]
    if funs!=None:
        fList=workFunctionList(funs)
    #print(fList)   
    
    
    rstO["name"]=clzName
    rstO["method"]=fList
    return rstO


def workClassList(classList):
    rst=[]
    for classO in classList:
        rst.append(parseClassO(classO))
        
        
    return rst
        
def getStructDataFromHtmlO(xmlO):
    rstO={}
    
    functionList=xmlO.find_all(None,"symbol function ")
    print(len(functionList))
    flist=workFunctionList(functionList)
    
    classList=xmlO.find_all(None,"symbol class")
    print(len(classList))
    clzList=workClassList(classList)
    rstO["functionList"]=flist
    rstO["classList"]=clzList
    return rstO
 
def workHtmPage(patePath,fileName):
    xmlO=readXml(patePath)
    dataO=getStructDataFromHtmlO(xmlO)
    #print(dataO)
    
    fw=open("tensorflowDes"+".json","w",encoding="utf-8")
    json.dump(dataO,fw,ensure_ascii=False,indent=4)

def getDirFiles(dirPath):
    files=os.listdir(dirPath)
    print(files)
    for tfile in files:
        tfile=dirPath+"/"+tfile
        #print("isdir:",os.path.isdir(tfile))
        if os.path.isdir(tfile):
            continue
        #print(tfile)
        bname=os.path.basename(os.path.realpath(tfile))
       # print(bname)
        names=bname.split(".")
        names.pop()
        names=names[0].split(" _ ")
        names.pop(0)
        #print(names)
        #print("_".join(names))
        workHtmPage(tfile,"_".join(names))
#workHtmPage("wiki/OrzWiki _ 学习 _ 量化交易.htm")

getDirFiles("APIHtml")
print("hihi")
```
